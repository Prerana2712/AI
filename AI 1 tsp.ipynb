```python
import itertools

num_nodes = int(input("Enter Number of Vertices :"))
num_edges = int(input("Enter Number of edges :"))

mat = [[0 for i in range(num_nodes)] for j in range(num_nodes)]

for i in range(num_edges):
    first_node = int(input("\nEnter Source Vertex :"))
    second_node = int(input("\nEnter Destination Vertex :"))
    weight = int(input("\nEnter Weight of this edge :"))
    mat[first_node][second_node] = weight
    mat[second_node][first_node] = weight

print("\n\n")
for i in range(num_nodes):
    for j in range(num_nodes):
        print(mat[i][j], end=" ")
    print()
print("\n\n")

def TSP_Implement(Adj_matrix, s):
    cities = []
    for i in range(num_nodes):
        if i != s:
            cities.append(i)
    min_distance = float('inf')
    print("\nPossible paths Cost \n")
    for perm in itertools.permutations(cities):
        curr_distance = 0
        k = s
        print(k, end="-->")
        for i in range(len(perm)):
            curr_distance += Adj_matrix[k][perm[i]]
            k = perm[i]
            print(k, end="-->")
        curr_distance += Adj_matrix[k][s]
        print("  ", curr_distance, " ")
        min_distance = min(min_distance, curr_distance)
        print("\n")
    return min_distance

print("\n\nMinimum Distance: ", TSP_Implement(mat, 0))
```
